```python
print("loading labels ....")
labels = pd.read_csv('../input/employeeattritiondataset/labels.csv', header=None)
Y = labels.values
print("loading features ....")
features = pd.read_csv('../input/employeeattritiondataset/features.csv', header=None)
X = features.values
```

```python
# split the data into training and testing sets
print("Splitting Datasets into Training and Testing Datasets")
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.4, random_state=17)
```

# K-Nearest Neighbors Classifier

```python
# KNN classifier
KNN = KNeighborsClassifier(n_neighbors=3, p=1)
# train the classifier
KNN.fit(x_train, y_train.ravel())
# predict the test samples
KNN_pred = KNN.predict(x_test)
# metrics evaluation
KNN_accuracy = (accuracy_score(y_test, KNN_pred))
KNN_Precision=precision_score(y_test, KNN_pred, average='macro') 
KNN_Recall = recall_score(y_test, KNN_pred, average='macro')
KNN_F1 = f1_score(y_test, KNN_pred, average='macro')

print("Accuracy :",KNN_accuracy,"\n%s" % 
      "Precision:",KNN_Precision,"\n%s" % 
      "Recall   :",KNN_Recall, "\n%s" % 
      "F1       :",KNN_F1, "\n")

print("Confusion Matrix:\n\n%s\n" % (confusion_matrix(y_test, KNN_pred)))
print("Classification Report:\n\n%s\n" % classification_report(y_test, KNN_pred, digits=5))

```

# Multi-layer Perceptron classifier

```python
# create a Multi-layer Perceptron Classifier
MLP =  MLPClassifier(activation='relu',max_iter=20, solver='adam')
# train the classifier
MLP.fit(x_train, y_train.ravel())
# predict the test samples
MLP_pred = MLP.predict(x_test)
# metrics evaluation
MLP_accuracy = (accuracy_score(y_test, MLP_pred))
MLP_Precision=precision_score(y_test, MLP_pred, average='macro') 
MLP_Recall = recall_score(y_test, MLP_pred, average='macro')
MLP_F1 = f1_score(y_test, MLP_pred, average='macro')

print("Accuracy :",MLP_accuracy,"\n%s" % 
      "Precision:",MLP_Precision,"\n%s" % 
      "Recall   :",MLP_Recall, "\n%s" % 
      "F1       :",MLP_F1, "\n")

print("Confusion Matrix:\n\n%s\n" % (confusion_matrix(y_test, MLP_pred)))
print("Classification Report:\n\n%s\n" % classification_report(y_test, MLP_pred, digits=5))

warnings.filterwarnings("ignore")
```
